```python
# 各科目の点数を変数として定義
japanese = 85 # 国語
math = 90     # 数学
english = 75  # 英語

# 平均点を計算
average = (japanese + math + english) / 3

# 結果を表示
print("3科目の平均点は", average, "点です")
```

```python
# 比較演算
a = 5
b = 10

#aがbより小さいかどうかを比較
result = a < b # これは True

# aとbが等しいかどうかを比較
equal_result = a == b # これは False

print("a < b:", result)
print("a == b:", equal_result)
```

```python
# ブール値に変換される値の例
print(bool(0)) # False
print(bool(1)) # True
print(bool("")) # False(空の文字列はFalse)
print(bool("Hello")) # True(非空の文字列はTrue)
```

```python
# True は 1 として扱われ、False は 0 として扱われる
print(True + True) # 1 + 1 = 2
print(True * 10)   # 1 * 10 = 10
print(False + 5)   # 0 + 5 = 5
```

```python
value = None # 変数を初期化
if value is None:
    print("まだ値が設定されていません")
```

```python
# 関数の返り値がない場合
def greet():
    print("Hello!")

result = greet() # 関数がNoneを返す
print(result)    # None と出力される
```
