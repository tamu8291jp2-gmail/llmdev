```python
# 課題：数値計算の例外処理を行おう
try:
    num1 = int(input("1つ目の数値: "))
    num2 = int(input("2つ目の数値: "))

    # 割り算を行う前にチェック
    if num2 == 0:
        raise ZeroDivisionError("エラー: 0で割ることはできません")
    
    # 安全な場合のみ割り算を行う
    result = num1 / num2
except ValueError:
    print("エラー: 数値を入力してください")
except ZeroDivisionError as e:
    print(e)
else:
    print(f"計算結果: {result}")
```
