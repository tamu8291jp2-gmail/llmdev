```python
# 料込価格を計算しよう
price_without_tax = int(input("税抜価格を入力してください: "))
price_with_tax = price_without_tax * 1.1

if price_with_tax >= 2000:
    print("送料は無料です")
    shipping_fee = 0
else:
    print("送料は350円です")
    shipping_fee = 350

total_price = price_with_tax + shipping_fee
print(f"送料込税込価格は{total_price:.0f}円です")
```
