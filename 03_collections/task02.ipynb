```python
# 課題：単語の文字数をカウントしよう
word_list = []
alphabet_counts = {'a':0, 'b':0, 'c':0, 'd':0, 'e':0, 'f':0, 'g':0, 'h':0, 'i':0, 'j':0, 'k':0, 'l':0, 'm':0, 'n':0, 'o':0, 'p':0, 'q':0, 'r':0, 's':0, 't':0, 'u':0, 'v':0, 'w':0, 'x':0, 'y':0, 'z':0}

while True:
    input_word = input("英単語を入力してください：")
    if input_word == "":
        break
    word_list.append(input_word)

    lower_word = input_word.lower()
    for char in lower_word:
        if char in alphabet_counts:
            alphabet_counts[char] += 1

word_list.sort()
print(f"入力した英単語：{word_list}")

for char, count in alphabet_counts.items():
    if count > 0:
        print(f"{char}が{count}個ありました")
```
