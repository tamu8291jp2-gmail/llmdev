```python
# 文字列の書き込み
with open("sample.txt", "w") as file:
    file.write("Hello, World!")
```

```python
# 複数行の書き込み
lines = ["First line\n", "Second line\n", "Third line\n"]
with open("sample.txt", "w") as file:
    file.writelines(lines)
```

```python
# テキストファイルの読み込み
with open("sample.txt", "r") as file:
    data = file.read()
    print(data)
```

```python
# 1行ずつ読み込み
with open("sample.txt", "r") as file:
    lines = file.readlines()
    for index, line in enumerate(lines):
        print(f"{index}:{line.strip()}")
```

```python
# 逐次行の読み込み
with open("sample.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
# CSVファイルへの書き込み

import csv

data = [
    ["name", "age", "city"],
    ["Alice", "30", "New York"],
    ["Bob", "25", "Los Angeles"],
    ["Charlie", "35", "Chicago"]
]

with open("sample1.csv", mode="w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)
```

```python
# 辞書をCSVファイルに書き込む

import csv

data = [
    {"name": "Alice", "age": "30", "city": "New York"},
    {"name": "Bob", "age": "25", "city": "Los Angeles"},
    {"name": "Charlie", "age": "35", "city": "Chicago"}
]

with open("sample2.csv", mode="w", encoding="utf-8", newline="") as file:
    fieldnames = ["name", "age", "city"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(data)
```

```python
# CSVファイルの読み込み

import csv

with open("sample1.csv", mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
# 辞書にCSVファイルを読み込む

import csv

with open("sample2.csv", mode="r", encoding="utf-8") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)
```

```python
# JSONファイルへの書き込み

import json

data = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "hobbies": ["reading", "traveling", "coding"]
}

with open("data.json", mode="w", encoding="utf-8") as file:
    json.dump(data, file, indent=4)
```

```python
# JSONファイルの読み込み

import json

with open("data.json", mode="r", encoding="utf-8") as file:
    data = json.load(file)
    print(data)
    print(type(data))
    print(data["name"])
```
